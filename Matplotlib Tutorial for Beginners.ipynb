```python
import numpy as np
import pandas as pd
```

```python
#import Matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
x1 = np.linspace(0,10,100)

#create a plot figure
fig = plt.figure()

plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'-');

```

```python
#crate a plot figure 
plt.figure

#create a first of two pannels to set current axis
plt.subplot(2,1,1)
plt.plot(x1,np.sin(x1))
```

```python
#create a first of two pannel to set current axis
plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1));
```

```python
#get current figure information
print(plt.gcf())
```

```python
#get current figure information
print(plt.gca())
```

```python
#Visualization with pyplot
plt.plot([1,2,3,4])
plt.ylabel('numbers')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
#State machine interface
x = np.linspace(0,2,100)
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('Simple Plot')
plt.legend()
plt.show()
```

```python
#Formatting the style of plot
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.show()
```

```python
#Evenly sampled times 200ms intervals
t=np.arange(0.,5.,0.2)
#red dashes blue squares green triangle
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
#First create a grid of plots
#ax will be an array of two axes objects
fig,ax = plt.subplots(2)
#call plot() method on the appropriate objects
ax[0].plot(x1,np.sin(x1),'b-')
ax[1].plot(x1,np.cos(x1),'b-')
```

```python
fig = plt.figure()
x2 = np.linspace(0,5,10)
y2 = x2**2
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x2,y2,'r')
axes.set_label('x2')
axes.set_label('y2')
axes.set_label('title')
```

```python
x3=np.arange(0.0,0.6,0.01)
plt.plot(x3,[xi**2 for xi in x3],'b-')
plt.show()
```

```python
#Multiline plots
x4 = range(1,5)
plt.plot(x4,[xi*1.5 for xi in x4])
plt.plot(x4,[xi*3 for xi in x4])
plt.plot(x4,[xi/3 for xi in x4])
plt.show()
```

```python
#Line plot
#create figure and axes first
fig = plt.figure()
ax=plt.axes()
#declare a variable x5
x5=np.linspace(0,10,1000)
#plot the sinousoid function
ax.plot(x5,np.sin(x5),'b-');
```

```python
# scatter plot
x7 = np.linspace (0,10,30)
y7 = np.sin(x7)
plt.plot(x7,y7,'o',color ='black');
```

```python
#Bar chart
data2 = (5.,25.,50.,20.)
plt.bar(range(len(data2)),data2)
plt.show()
```

```python
#Horizontal bar chart
data2 = (5.,25.,50.,20.)
plt.barh(range(len(data2)),data2)
plt.show()
```

```python
#Error bar chart
x9=np.arange(0,4,0.2)
y9=np.exp(-x9)
e1=0.1*np.abs(np.random.randn(len(y9)))
plt.errorbar(x9,y9,yerr=e1,fmt='.-')
plt.show()
```

```python
#Stacked bar chart
A =(15.,30.,45.,22.)
B =(15.,25.,50.,20.)
z2=range(4)
plt.bar(z2,A,color='b')
plt.bar(z2,B,color ='r',bottom =A)
plt.show()
```

```python
#Pie chart
plt.figure(figsize=(7,7))
x10=(25,35,20,20)
labels=['Computer','Chemical','Mechanical','Electronics']
plt.pie(x10, labels=labels);
plt.show()
```

```python
#Box plot
data3 = np.random.randn(100)
plt.boxplot(data3)
plt.show();
```

```python
#Area chart
x12=range(1,6)
y12=(1,4,6,8,4)
#Area plot
plt.fill_between(x12,y12)
plt.show()
```

```python
#Contour plot
#Create matrix
matrix1=np.random.randn(10,20)
cp=plt.contour(matrix1)
plt.show()
```

```python
#Adding a grid 
x15=np.arange(1,5)
plt.plot(x15*1.5,x15*3.0,x15,x15/3.0)
plt.grid(True)
plt.show()
```

```python
#Handling axes 
x15=np.arange(1,5)
plt.plot(x15,x15*1.5,x15,x15*3.0,x15,x15/3.0)
plt.axis()
plt.axis([0,5,-1,13])
plt.show()
```

```python
x15=np.arange(1,5)
plt.plot(x15,x15*1.0,x15,x15*3.0,x15,x15/3.0)
plt.xlim(1.0,4.0)
plt.ylim(0.0,12.0)
plt.show()
```

```python
#Handling x and y tricks 
u=[5,4,9,7,8,9,6,5,7,8]
plt.plot(u)
plt.xticks([2,4,6,8,10])
plt.yticks([2,4,6,8,10])
plt.show()
```

```python
#Adding labels
plt.plot([1,2,3,4])
plt.xlabel('This is the X is axis')
plt.ylabel('This is the Y is axis')
plt.show()
```

```python
#Adding a titile
plt.plot([1,2,3,4])
plt.title('First Plot')
plt.show()
```

```python
#Adding a legend
x15= np.arange(1,5)
fig,ax=plt.subplots()
ax.plot(x15,x15*1.5)
ax.plot(x15,x15*3.0)
ax.plot(x15,x15/3.0)
ax.legend(['Normal','Fast','Slow'])
```

```python
x15= np.arange(1,5)
fig,ax=plt.subplots()
ax.plot(x15,x15*1.5,label = 'Normal')
ax.plot(x15,x15*3.0,label = 'Fast')
ax.plot(x15,x15/3.0,label = 'Slow')
ax.legend()
```

```python
#Control colours
x16=np.arange(1.5)
plt.plot(x16,'r')
plt.plot(x16+1,'g')
plt.plot(x16+2,'b')
plt.show()
```

```python
#Control line styles
x16=np.arange(1,5)
plt.plot(x16,'--',x16+1,'-.',x16+2,':')
plt.show()
```
